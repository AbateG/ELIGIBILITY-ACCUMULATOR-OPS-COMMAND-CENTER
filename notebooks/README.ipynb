# Notebooks

This folder contains optional analytical companion notebooks for the **Eligibility & Accumulator Operations Command Center**.

## Why Notebooks exist in this project?

The primary experience of this project is delivered through:

- The Streamlit application

- SQL Query Workbench

- Investigation Playbooks

- Scenario Runbooks

- SQL Playbooks

- Automated Tests

These notebooks are intentionally supplemental. They are designed to provide:

- A fast analytical tour of the database

- Validation of deterministic operational scenarios

- Deeper explanation of accumulator business logic

They are **not** intended to replace the application or duplicate every dashboard page.

## Notebook Inventory

### 1. 'O1_project_data_tour.ipynb'

A guided introduction to the project's schema, seeded data, major operational entities, and implemented scenarios.

### 2. '02_sceario_validation_workbook.ipynb'

A validation-focused notebook confirming that each deterministic scenario is properly represented across issues, support cases, SLAs, files, runs, members, and claims.

### 3. '03_accumulator_logic_deep_dive.ipynb'

A domain-focused notebook explaining how claims, out-of-pocket accumulation, and family rollups work in the simulator.
